```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
path = "/content/drive/MyDrive/ML/baby-weights-dataset.csv"
df = pd.read_csv(path)
```

```python
df = df[['WEEKS', 'BWEIGHT']].dropna()
```

```python
age = df['WEEKS'].values
birthwt = df['BWEIGHT'].values
```

```python
age = age.reshape(-1, 1)
```

```python
lr = LinearRegression()
```

```python
lr.fit(age, birthwt)
```

```python
y_pred = lr.predict(age)
```

```python
print("Coefficient:", lr.coef_[0])
print("Intercept:", lr.intercept_)
```

```python
plt.scatter(age, birthwt)
plt.plot(age, y_pred)
plt.xlabel("Gestation Age (Weeks)")
plt.ylabel("Birth Weight")
plt.title("Linear Regression: Baby Weight vs Gestation Age")
plt.show()
```
